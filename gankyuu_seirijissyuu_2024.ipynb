# **Physiological Nystagmus 生理的な眼球振盪**

Over long periods of stimulation evoking the opto-kinetic reflex (OKR) or vestibulo-ocular reflex (VOR), the eyes exhibit a stereotypical pattern of alternating *slow phases* and *quick phases* (also called *fast phases*). This is called *nystagmus*. The nystagmus elicited by optokinetic stimulation is called optokinetic nystagmus (OKN). Vestibulo-ocular stimulation evokes vestibulo-ocular nystagmus (VON; this abbreviation is not common).

* During the **slow phase** of OKN/VON, the reflex causes the eyes to smoothly track the visual stimulus (OKR) or counter-rotate against the head rotation (VOR).

* During the **quick phases** of OKN/VON, the eyes make (probably involuntary) saccade-like movements in the **opposite direction to the slow phase**. The purpose is to re-center or advance the eye ahead of the perceived (visual or vestibular) motion.

---

The OKR/VOR slow phases operate on every axis of eye rotation (horizontal, vertical, and torsional). Typical (video) eye tracking systems can only detect horizontal and vertical components.

# **Measurement of eye movements**

There are several ways to measure eye movements. The most common are:

眼球運動計測方法は主に以下の３つ。

1. Electro-oculography (EOG)
2. Scleral Search Coils
3. Video-oculography (VOG)

In this lab, we will use video-oculography to observe eye movements. We illuminate the eye using an infrared illuminator, and calculate the angle of the eyeball using (1) the distance from the center of the pupil to the reflection on the cornea (corneal reflection, CR, a.k.a. "glint"), and (2) the relative angle between the CR and the pupil center. The pupil is relatively easy to detect because it is a hole in the eyeball from which light does not return, thus becoming an ellipsoid-shaped "shadow" in the infrared video of the eye.

We illuminate the eye using infrared light (~900 nm). An infrared camera captures the eyes and detects the pupil (=hole) and the corneal reflection (CR - the reflection of the illuminating light off the cornea of the lens). We calculate the distance between the reflection and the center of the pupil (in pixels on the eye camera image) and compare it to the distance and direction at known orientations of the eyeball, i.e. when the subject is looking at the calibration dots on the screen). We use this to fit a 2d (or more) polynomial model, which predicts screen position when given a new (unknown) eye camera position. By knowing the screen distance and the width of a pixel in meters, we can calculate the angle using the tangent function.

本実習では、Video-oculography をもって眼球運動の観察をする。赤外線イルミネーターで眼球を照らし、角膜においてのイルミネーターのまぶしい反射(CR)と瞳孔部分の中心(pupil center)との距離、そして方向に基づき、眼球の傾きを計算する。

瞳孔は眼球の球体にあいている「穴」であるため、光が眼球の中に入ったら戻ってこない楕円の「影」になり、感知しやすい。
